# Spatial Dispersion: Standard Deviation Ellipse dan Standard Distance

## 1. Konsep Spatial Dispersion
Spatial Dispersion (penyebaran spasial) adalah ukuran statistik yang digunakan untuk memahami seberapa jauh, merata, atau terkonsentrasi suatu fenomena di ruang geografis. Jika *spatial central tendency* menjawab pertanyaan “dimana pusat distribusi?”, maka *spatial dispersion* menjawab pertanyaan: **“seberapa tersebar titik-titik di sekitar pusat tersebut, dan dengan arah apa?”**

Dua metode yang umum digunakan dalam GIS:
- **Standard Distance (SD)**
- **Standard Deviation Ellipse (SDE)**

---
## 2. Standard Distance (SD)
**Definisi:**
Standard Distance adalah ukuran seberapa jauh rata-rata titik (centroid/mean center) terhadap seluruh distribusi titik. Konsep ini mirip dengan *standard deviation* dalam statistik biasa, tetapi diterapkan pada data spasial.

**Rumus:**
\[ SD = \sqrt{\frac{\sum (x_i - \bar{x})^2 + (y_i - \bar{y})^2}{n}} \]
dengan:
- \(x_i, y_i\) = koordinat titik ke-i
- \(\bar{x}, \bar{y}\) = mean center
- \(n\) = jumlah titik

**Interpretasi:**
- SD menghasilkan lingkaran konsentris dengan radius yang merepresentasikan satu *standard deviation* dari pusat.
- Jika titik tersebar luas, radius SD besar. Jika terkonsentrasi, radius kecil.

**Kegunaan:**
- Analisis kejahatan: mengukur seberapa terkonsentrasi insiden kriminal.
- Perencanaan fasilitas: mengetahui seberapa jauh warga tersebar dari pusat layanan.
- Ekologi: memahami sebaran spesies dalam habitat.

---
## 3. Standard Deviation Ellipse (SDE)
**Definisi:**
SDE adalah pengembangan dari Standard Distance. Tidak hanya menghitung jarak sebaran, tetapi juga arah dominan penyebaran. Bentuk hasilnya berupa **ellips**.

**Karakteristik:**
- Memiliki tiga parameter utama:
  1. **Mean center** (pusat ellips)
  2. **Arah utama (azimuth/rotation)** – orientasi ellips
  3. **Panjang sumbu mayor dan minor** – variasi sebaran sepanjang arah utama dan tegak lurusnya

**Rumus komponen orientasi:**
\[ \tan(2\theta) = \frac{2\sum (x_i-\bar{x})(y_i-\bar{y})}{\sum (x_i-\bar{x})^2 - \sum (y_i-\bar{y})^2} \]
dimana \(\theta\) adalah arah utama penyebaran.

**Interpretasi:**
- Jika distribusi mendekati lingkaran → sebaran isotropik (tanpa arah dominan).
- Jika ellips memanjang → ada pola orientasi arah dominan.

**Kegunaan:**
- Analisis transportasi: memahami orientasi jaringan jalan/arus lalu lintas.
- Analisis kriminal: mendeteksi arah utama persebaran tindak kriminal.
- Studi geomorfologi: mempelajari arah dominan lembah sungai.
- Urban planning: mendeteksi arah pertumbuhan kota (sprawl direction).

---
## 4. Perbandingan SD dan SDE
| Aspek                  | Standard Distance (SD)           | Standard Deviation Ellipse (SDE) |
|------------------------|---------------------------------|----------------------------------|
| Bentuk hasil           | Lingkaran                      | Ellips                           |
| Informasi arah         | Tidak ada                      | Ada (orientasi dominan)           |
| Sumbu utama            | Tidak ada                      | Mayor dan minor                   |
| Aplikasi umum          | Konsentrasi spasial            | Konsentrasi + arah distribusi     |

---
## 5. Aplikasi dalam Geografi dan Perencanaan Wilayah
1. **Geografi Fisik**
   - Pola distribusi spesies flora/fauna.
   - Pola erosi dan sedimentasi.

2. **Geografi Manusia**
   - Distribusi permukiman.
   - Analisis pusat-pusat aktivitas ekonomi.

3. **Perencanaan Wilayah dan Kota**
   - Mengidentifikasi arah pertumbuhan kota (urban sprawl).
   - Perencanaan transportasi → menentukan arah jalur angkutan massal.
   - Menentukan lokasi fasilitas publik (rumah sakit, sekolah, halte bus) yang optimal berdasarkan penyebaran penduduk.

4. **Kriminologi**
   - Memahami pola spasial kejahatan dan orientasi pergerakannya.

5. **Kesehatan Masyarakat**
   - Penyebaran penyakit menular.

---
## 6. Kesimpulan
- **Standard Distance** menjawab pertanyaan seberapa jauh penyebaran.
- **Standard Deviation Ellipse** menjawab pertanyaan seberapa jauh **dan** ke arah mana penyebaran.
- Keduanya penting dalam analisis spasial untuk memberikan gambaran kuantitatif dan visual tentang pola distribusi.
